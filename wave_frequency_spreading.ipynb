```python
import numpy as np
import matplotlib.pyplot as plt
%run -i runup_funcs.py

H0 = 2.
Tp = 10.
```

```python
fp = 1./Tp
L0 = calc_L0(Tp)
Beta = 0.02

nfs = 8
fs = np.linspace( 0.0025, 0.02, num=nfs)
nsw = 26
sw = np.linspace( 5., 30, num=nsw)

RigGF12 = np.zeros((len(fs),len(sw)))
                   
for i, fsi in enumerate(fs):
    RigGF12[i,:] = calc_RigGF12(H0, L0, Beta, fp, fsi, sw  )
```

```python
S = .06*np.sqrt(H0*L0)
```

```python
plt.plot(sw,S*np.ones_like(sw),'--',c='gray')
for i in range( nfs ):
    ts = '{:.1f}'.format(fp/fs[i])
    plt.plot(sw,RigGF12[i,:],label=ts)
plt.legend(bbox_to_anchor=(.85, .5))
#plt.legend(loc='upper right')
plt.xlabel(r'Directional Spreading ($^{\circ}$)', fontsize=16)
plt.ylabel(r'Runup $R_{ig}$ (m)', fontsize=16)

plt.savefig('GF_IG_effect.svg',dpi=200)
```
